# Preprocessing

```python
def window(df, window_size=60):
    def split_df(df, start, end):
        split = df.iloc[start:end]
        X = split.iloc[:, :-1]
        y = split.iloc[:, -1]
        return X, y

    def window_data(X, y, window_size):
        X_windows = [X[i:i + window_size] for i in range(len(X) - window_size)]
        y_windows = [y[i + window_size] for i in range(len(X) - window_size)]
        return np.array(X_windows), np.array(y_windows)

    # Split the data
    train_size = int(len(df) * 0.7)
    val_size = int(len(df) * 0.15)
    X_train_df, y_train_df = split_df(df, 0, train_size)
    X_val_df, y_val_df = split_df(df, train_size, train_size + val_size)
    X_test_df, y_test_df = split_df(df, train_size + val_size, None)

    # Initialize the scaler
    sc = MinMaxScaler(feature_range=(0,1))
    X_train = sc.fit_transform(X_train_df).astype('float')
    X_val = sc.transform(X_val_df).astype('float')
    X_test = sc.transform(X_test_df).astype('float')
    
    y_train = y_train_df.to_numpy().astype('float')
    y_val = y_val_df.to_numpy().astype('float')
    y_test = y_test_df.to_numpy().astype('float')

    # Create windows
    X_train_windows, y_train_windows = window_data(X_train, y_train, window_size)
    X_val_windows, y_val_windows = window_data(X_val, y_val, window_size)
    X_test_windows, y_test_windows = window_data(X_test, y_test, window_size)

    return (X_train_windows, y_train_windows, X_val_windows, y_val_windows, X_test_windows, y_test_windows,
           X_train_df, y_train_df, X_val_df, y_val_df, X_test_df, y_test_df)
```

```python
(aapl_X_train, aapl_y_train, aapl_X_val, aapl_y_val, aapl_X_test, aapl_y_test,
 aapl_X_train_df, aapl_y_train_df, aapl_X_val_df, aapl_y_val_df,
 aapl_X_test_df, aapl_y_test_df) = window(aapl)

(intc_X_train, intc_y_train, intc_X_val, intc_y_val, intc_X_test, intc_y_test,
 intc_X_train_df, intc_y_train_df, intc_X_val_df, intc_y_val_df,
 intc_X_test_df, intc_y_test_df) = window(intc)
```

```python
def load_models_and_scaler(symbol):
    base_path = f'/home/adedapo/personal_project/daps05ayoade/disseration/alternative-models/'
    
    # Load models
    cnn = load_model(base_path + f'best_cnn_model_{symbol}.h5')
    rnn = load_model(base_path + f'best_rnn_model_{symbol}.h5')
    lstm = load_model(base_path + f'best_lstm_model_{symbol}.h5')
    gru = load_model(base_path + f'best_gru_model_{symbol}.h5')
    
    # Load scaler
    with open(f'/home/adedapo/personal_project/daps05ayoade/disseration/alternative-scaler/scaler_{symbol}.pkl', 'rb') as file:
        scaler = pickle.load(file)
        
    return cnn, rnn, lstm, gru, scaler
```

```python
def evaluate_and_predict(df, cnn, rnn, lstm, gru, scaler):
    data_fetcher = DataFetcher
    stock_data = df.drop(columns='target')
    test = df['adj_close']
    
    scaled_stock_data = scaler.transform(stock_data)
    scaled_windows = data_fetcher.window_data(scaled_stock_data, window_size=60)
    
    min_length = min(len(scaled_windows), len(test))
    scaled_windows_trimmed = scaled_windows[:min_length]
    test_trimmed = test[:min_length]

    predictions = {
        "rnn": rnn.predict(scaled_windows_trimmed),
        "cnn": cnn.predict(scaled_windows_trimmed),
        "lstm": lstm.predict(scaled_windows_trimmed),
        "gru": gru.predict(scaled_windows_trimmed)
    }

    losses = {
        "rnn": rnn.evaluate(scaled_windows_trimmed, test_trimmed),
        "cnn": cnn.evaluate(scaled_windows_trimmed, test_trimmed),
        "lstm": lstm.evaluate(scaled_windows_trimmed, test_trimmed),
        "gru": gru.evaluate(scaled_windows_trimmed, test_trimmed)
    }

    for model_name, loss in losses.items():
        print(f'{model_name.upper()} Test loss:', loss)

    return test_trimmed, predictions
```

```python
def calculate_accuracy(y_true, y_pred, tolerance=0.01):
    """Calculate the accuracy based on a given tolerance range."""
    # Convert to numpy arrays if they are pandas data structures
    if isinstance(y_true, (pd.Series, pd.DataFrame)):
        y_true = y_true.values
    if isinstance(y_pred, (pd.Series, pd.DataFrame)):
        y_pred = y_pred.values

    # Squeeze y_pred to ensure it's 1-dimensional
    y_pred = np.squeeze(y_pred)

    # Check for shape consistency
    assert y_true.shape == y_pred.shape, "Shape mismatch between true values and predictions"

    return np.mean(np.abs((y_pred - y_true) / y_true) < tolerance)


def evaluate_and_predict(df, cnn, rnn, lstm, gru, scaler):
    data_fetcher = DataFetcher
    stock_data = df.drop(columns='target')
    test = df['adj_close']
    
    scaled_stock_data = scaler.transform(stock_data)
    scaled_windows = data_fetcher.window_data(scaled_stock_data, window_size=60)
    
    min_length = min(len(scaled_windows), len(test))
    scaled_windows_trimmed = scaled_windows[:min_length]
    test_trimmed = test[:min_length]

    predictions = {
        "rnn": rnn.predict(scaled_windows_trimmed),
        "cnn": cnn.predict(scaled_windows_trimmed),
        "lstm": lstm.predict(scaled_windows_trimmed),
        "gru": gru.predict(scaled_windows_trimmed)
    }

    losses = {
        "rnn": rnn.evaluate(scaled_windows_trimmed, test_trimmed),
        "cnn": cnn.evaluate(scaled_windows_trimmed, test_trimmed),
        "lstm": lstm.evaluate(scaled_windows_trimmed, test_trimmed),
        "gru": gru.evaluate(scaled_windows_trimmed, test_trimmed)
    }

    accuracies = {
        "rnn": calculate_accuracy(test_trimmed, predictions["rnn"]),
        "cnn": calculate_accuracy(test_trimmed, predictions["cnn"]),
        "lstm": calculate_accuracy(test_trimmed, predictions["lstm"]),
        "gru": calculate_accuracy(test_trimmed, predictions["gru"])
    }

    for model_name, loss in losses.items():
        print(f'{model_name.upper()} Test loss:', loss)
        print(f'{model_name.upper()} Accuracy:', accuracies[model_name] * 100, '%')

    return test_trimmed, predictions
```

```python
MODEL_PATH_ROOT = '/home/adedapo/personal_project/daps05ayoade/disseration/trained-model/'
SCALER_PATH_ROOT = '/home/adedapo/personal_project/daps05ayoade/disseration/fitted-scaler/'
DATA_PATH = '/home/adedapo/personal_project/daps05ayoade/disseration/data/200_large_cap_stocks.csv'

data_fetcher = DataFetcher(API_TOKEN)
```

```python
symbols_df = pd.read_csv(DATA_PATH)
tickers = symbols_df['Ticker'].tolist()
```

```python
def fetch_data_for_ticker(ticker):
    return get_stock_data(ticker)

def prepare_and_evaluate(ticker, ticker_data):
    # Prepare paths for the model and scaler
    model_path = f'{MODEL_PATH_ROOT}best_model_{ticker}.h5'
    scaler_path = f'{SCALER_PATH_ROOT}scaler_{ticker}.pkl'
    
    # Load model and scaler
    model = load_model(model_path)
    with open(scaler_path, 'rb') as f:
        scaler = pickle.load(f)
    
    # Fetch data for the specific ticker
    data = ticker_data.drop(columns='target')
    ytest = ticker_data['adj_close']
    
    # Scale the data
    scaled_data = scaler.transform(data)
    
    # Window the data (assuming data_fetcher.window_data is defined elsewhere)
    sc_windows = data_fetcher.window_data(scaled_data, window_size=60)
    
    # Trim data
    min_length = min(len(sc_windows), len(ytest))
    sc_windows_trimmed = sc_windows[:min_length]
    ytest_trimmed = ytest[:min_length]
    
    # Evaluate the model (assuming evaluate is a function of the model)
    scores = model.evaluate(sc_windows_trimmed, ytest_trimmed)
    
    return scores

# Iterate through each ticker, fetch data, and evaluate
all_scores = {}
counter = 0
for ticker in tqdm(tickers, desc="Processing tickers"):
    # Fetch data for the ticker
    ticker_data = fetch_data_for_ticker(ticker)
    
    # Evaluate the model for the ticker
    all_scores[ticker] = prepare_and_evaluate(ticker, ticker_data)
    
    # Increment the counter
    counter += 1
    
    # Sleep for 60 seconds after processing 11 tickers
    if counter % 11 == 0:
        time.sleep(60)
```

# Sentiment Performance

```python
def fetch_sentiment_data_for_all_tickers(csv_path, api_token):
    # 1. Load the tickers from the CSV
    tickers_df = pd.read_csv(csv_path)
    tickers = tickers_df['Ticker'].tolist()  # Assuming the column name with tickers is 'ticker'

    all_ticker_data = []

    # 2. Loop through each ticker
    for ticker in tickers:
        try:
            url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&apikey={api_token}'
            response = requests.get(url)
            data = response.json()

            # Check if the response contains an error message
            if "Note" in data or "Error Message" in data:
                print(f"Error fetching data for {ticker}: {data.get('Note', data.get('Error Message', 'Unknown error'))}")
                continue

            # 3. Extract sentiment data for the current ticker
            for feed_item in data['feed']:
                for ticker_data in feed_item["ticker_sentiment"]:
                    if ticker_data['ticker'] == ticker:
                        ticker_item = {
                            'ticker': ticker,
                            'relevance_score': float(ticker_data['relevance_score']),
                            'ticker_sentiment_score': float(ticker_data['ticker_sentiment_score']),
                            'ticker_sentiment_label': ticker_data['ticker_sentiment_label']
                        }
                        all_ticker_data.append(ticker_item)

            # Sleep for 0.7 seconds to ensure you don't exceed 100 requests per minute
            time.sleep(0.7)

        except KeyError as e:
            print(f"KeyError encountered for {ticker}: {e}. Continuing to next ticker.")
        except Exception as e:
            print(f"Unexpected error for {ticker}: {e}. Continuing to next ticker.")

    return pd.DataFrame(all_ticker_data)
```

```python
url = "https://paper-api.alpaca.markets/v2/account/portfolio/history"

headers = {
    "accept": "application/json",
    "APCA-API-KEY-ID": API_KEY,
    "APCA-API-SECRET-KEY": SECRET_KEY
}

response = requests.get(url, headers=headers)
data = response.json()

# Extract data
timestamps = data["timestamp"]
equity = data["equity"]
profit_loss = [x for x in data['profit_loss'] if x is not None]
profit_loss = profit_loss[1:]
profit_loss_pct = [x*100 for x in data["profit_loss_pct"] if x is not None]
profit_loss_pct = profit_loss_pct[1:]

# Filter out null values and calculate metrics
equity = [x for x in equity if x is not None]
equity = equity[1:]
total_profit_loss = sum(profit_loss)
total_profit_loss_pct = sum(profit_loss_pct)

# Display metrics
print(f"Total Profit/Loss: ${total_profit_loss:.2f}")
print(f"Total Percentage Profit/Loss: {total_profit_loss_pct:.2f}%")
print(f"Starting Equity: ${equity[0]:.2f}")
print(f"Ending Equity: ${equity[-1]:.2f}")
```

```python
# Modify the stock_changes list to cover the new intervals
stock_changes = [7, 12, 17, len(profit_loss_pct)]

# Calculate total profit/loss for each period
profits_losses = []
start_index = 0
for end_index in stock_changes:
    period_data = profit_loss_pct[start_index:end_index]
    total_profit_loss = sum(period_data)
    profits_losses.append(total_profit_loss)
    start_index = end_index


print(f"Returns of 50 Stocks: {profits_losses[0]:.2f}%")
print(f"Returns of 100 Stocks: {profits_losses[1]:.2f}%")
print(f"Returns of 150 Stocks: {profits_losses[2]:.2f}%")
print(f"Returns of 200 Stocks: {profits_losses[3]:.2f}%")
```

```python
# 1. Volatility of Returns
# Calculate volatility for each period
volatilities = []
start_index = 0
for end_index in stock_changes:
    period_data = profit_loss_pct[start_index:end_index]
    volatility = np.std(period_data)
    volatilities.append(volatility)
    start_index = end_index
print(f"Volatiltiy of 50 Stocks: {volatilities[0]:.2f}%")
print(f"Volatiltiy of 100 Stocks: {volatilities[1]:.2f}%")
print(f"Volatiltiy of 150 Stocks: {volatilities[2]:.2f}%")
print(f"Volatiltiy of 200 Stocks: {volatilities[3]:.2f}%\n")

volatility = np.std(profit_loss_pct)
print(f"Volatility of Returns (Std Dev): {volatility:.2f}%")

# 2. Max Drawdown
rolling_max = np.maximum.accumulate(equity)  # Computes the running maximum
daily_drawdown = (equity/rolling_max) - 1.0
max_daily_drawdown = np.min(daily_drawdown)*100
print(f"\n Max Daily Drawdown: {max_daily_drawdown:.2f}%")
```

```python
# Given 3-month T-bill rate
Rf_3month = 0.0528

# Number of trading days in 3 months
n = 63

# Convert the 3-month T-bill rate to a daily rate using continuous compounding
R_f_daily = np.exp(Rf_3month / n) - 1

# Calculate the average daily return
R_p_daily = np.mean(profit_loss_pct)

# Calculate portfolio volatility (daily)
sigma_p_daily = np.std(profit_loss_pct)

# Compute Sharpe Ratio
sharpe_ratio = (R_p_daily - R_f_daily) / sigma_p_daily

print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```

## Daily Variations and Cumulative Returns Relative to SPY

```python
# Calculate cumulative returns
cumulative_returns = (np.array(profit_loss_pct) + 1).cumprod()

# Data Processing
SPY_url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=SPY&outputsize=compact&apikey={API_TOKEN}'

SPY = requests.get(SPY_url).json()
spy_df = pd.DataFrame(SPY['Time Series (Daily)']).T
spy_data = spy_df['5. adjusted close']

# Provided SPY adjusted close data
spy_data_values = spy_data.values

# Convert string values of SPY data to floats
spy_data_float = [float(val) for val in spy_data_values]

# Considering the length of the portfolio's active trading days
trading_days = len(cumulative_returns)

# Extract the most recent trading days for SPY corresponding to the portfolio's active days
spy_data_recent = spy_data_float[:trading_days]

# Calculate daily returns for the recent SPY data
spy_daily_returns_recent = [0]  # Starting with 0% for the first day
for i in range(1, len(spy_data_recent)):
    daily_return = (spy_data_recent[i] - spy_data_recent[i-1]) / spy_data_recent[i-1]
    spy_daily_returns_recent.append(daily_return)

# Compute cumulative returns for SPY based on daily returns
spy_cumulative_returns = (np.array(spy_daily_returns_recent) + 1).cumprod()

# Calculate cumulative returns
cumulative_returns = (np.array(profit_loss_pct) + 1).cumprod()
```
