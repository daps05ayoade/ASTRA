# Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler

def window(df, window_size):
    # Split the data into train, validation and test sets
    train_size = int(len(df) * 0.7)
    val_size = int(len(df) * 0.15)
    train_df, val_df, test_df = df.iloc[:train_size], df.iloc[train_size:train_size+val_size], df.iloc[train_size+val_size:]

    # Split into features and target for each set
    X_train_df = train_df.iloc[:, :-1]  # All columns except the last
    y_train_df = train_df.iloc[:, -1]  # Only the last column
    X_val_df = val_df.iloc[:, :-1]  # All columns except the last
    y_val_df = val_df.iloc[:, -1]  # Only the last column
    X_test_df = test_df.iloc[:, :-1]  # All columns except the last
    y_test_df = test_df.iloc[:, -1]  # Only the last column

    # Initialize the scaler
    sc = MinMaxScaler(feature_range=(0,1))

    # Fit the scaler on the training data and transform both training and testing data
    X_train = sc.fit_transform(X_train_df)
    X_val = sc.transform(X_val_df)
    X_test = sc.transform(X_test_df)

    # Convert the target data to numpy arrays
    y_train = y_train_df.to_numpy()
    y_val = y_val_df.to_numpy()
    y_test = y_test_df.to_numpy()

    X_train_windows = []
    y_train_windows = []
    X_val_windows = []
    y_val_windows = []
    X_test_windows = []
    y_test_windows = []

    for i in range(len(X_train) - window_size):
        X_train_windows.append(X_train[i:i + window_size])  # Exclude the last column (target)
        y_train_windows.append(y_train[i + window_size])  # The last column is the target

    for i in range(len(X_val) - window_size):
        X_val_windows.append(X_val[i:i + window_size])  # Exclude the last column (target)
        y_val_windows.append(y_val[i + window_size])  # The last column is the target

    for i in range(len(X_test) - window_size):
        X_test_windows.append(X_test[i:i + window_size])  # Exclude the last column (target)
        y_test_windows.append(y_test[i + window_size])  # The last column is the target

    return (np.array(X_train_windows), np.array(y_train_windows), np.array(X_val_windows), 
            np.array(y_val_windows), np.array(X_test_windows), np.array(y_test_windows), 
            train_df, val_df, test_df)

window_size = 60
X_train, y_train, X_val, y_val, X_test, y_test, train_df, val_df, test_df = window(df, window_size)

print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'X_val: {X_val.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
print(f'y_val: {y_val.shape}')
```

```python
# We have chosen 'High' attribute for prices. Let's see what it looks like
train_df['adj_close'].plot(figsize=(16,4),legend=True)
val_df['adj_close'].plot(figsize=(16,4), legend=True)
test_df['adj_close'].plot(figsize=(16,4),legend=True)
plt.legend(['Training set (70%)','Validation Set (15%)','Test set (15%)'])
plt.title('APPL stock price')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *

# Define the CNN model
model = Sequential()

# Add the first Conv1D layer
model.add(Conv1D(filters=128, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(MaxPooling1D(pool_size=2))

# Add a second Conv1D layer
model.add(Conv1D(filters=128, kernel_size=2, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

# Flatten the data for input into the Dense layer
model.add(Flatten())

# Add a Dense layer
model.add(Dense(50, activation='relu'))

# Output layer
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, batch_size=15, epochs=100, validation_data=(X_val, y_val))
```

### Evaluating CNN Model

```python
y_pred = model.predict(X_test)
result = pd.DataFrame(data={'Prediction': y_pred.flatten(), 'Actuals': y_test})
result.tail()
```

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test)
print('Test loss:', loss)

# Compute metrics
mae = np.mean(np.abs(y_test - y_pred))
rmse = np.sqrt(np.mean((y_test - y_pred)**2))

print('Mean Absolute Error:', mae)
print('Root Mean Squared Error:', rmse)
```

# Gridsearch CNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

def create_model(filters=64, kernel_size=2, activation='relu', optimizer='adam'):
    # Define the CNN model
    model = Sequential()

    # Add the first Conv1D layer
    model.add(Conv1D(filters=filters, kernel_size=kernel_size, activation=activation, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(MaxPooling1D(pool_size=2))

    # Add a second Conv1D layer
    model.add(Conv1D(filters=filters*2, kernel_size=kernel_size, activation=activation))
    model.add(MaxPooling1D(pool_size=2))

    # Flatten the data for input into the Dense layer
    model.add(Flatten())

    # Add a Dense layer
    model.add(Dense(50, activation=activation))

    # Output layer
    model.add(Dense(1))

    model.compile(optimizer=optimizer, loss='mse')

    return model

# Wrap Keras model with KerasRegressor
model = KerasRegressor(build_fn=create_model, epochs=10, batch_size=15, verbose=0)  # Use fewer epochs for the search

# Define the random search parameters
param_dist = {
    'filters': [32, 64, 128],
    'kernel_size': [2, 3, 5],
    'activation': ['relu', 'tanh'],
    'optimizer': ['SGD', 'Adam', 'RMSprop']
}

# Create Random Search
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=10, cv=3, random_state=42)
random_search_result = random_search.fit(X_train[:1000], y_train[:1000])  # Use a subset of your data for the search

# Summarize results
print("Best: %f using %s" % (random_search_result.best_score_, random_search_result.best_params_))
```
